```python
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()

        self.gradients = None

        self.conv1 = nn.Sequential(
        
          nn.Conv3d(1, 1, (3,3,3), groups=1, bias=False),
          #inplace_abn.InPlaceABN(1),
          nn.ReLU6(inplace=True),
          # pw-linear
          nn.Conv3d(1, 20, 1, 1, 0, bias=False),
          nn.ReLU(),
          nn.MaxPool3d((2, 2, 2)),
          
            )
        
        self.conv2 = nn.Sequential(
        
          nn.Conv3d(20, 20, (4,4,4), groups=20, bias=False),
          nn.ReLU6(inplace=True),
          # pw-linear
          nn.Conv3d(20, 40, 1, 1, 0, bias=False),
          nn.GroupNorm(5, 40),
          nn.ReLU(),
          nn.MaxPool3d((2, 2, 2)),
          
          
            )
        
        self.conv3 = nn.Sequential(
        
          nn.Conv3d(40, 40, (3,3,3), groups=40, bias=False),
          nn.ReLU6(inplace=True),
          # pw-linear
          nn.Conv3d(40, 80, 1, 1, 0, bias=False),
          nn.GroupNorm(10, 80),
          nn.ReLU(),
          nn.MaxPool3d((2, 2, 2)),
          
            )
        
        self.conv4 = nn.Sequential(
        
          nn.Conv3d(80, 80, (3,3,3), groups=80, bias=False),
          nn.ReLU6(inplace=True),
          # pw-linear
          nn.Conv3d(80, 160, 1, 1, 0, bias=False),
          nn.GroupNorm(20, 160),
          nn.ReLU(),
          
          
          nn.MaxPool3d((2, 2, 2)),
          
          
            )
        self.conv5 = nn.Sequential(
        
          nn.Conv3d(160, 160, (4,3,3), groups=160, bias=False),
          nn.ReLU6(inplace=True),
          # pw-linear
          nn.Conv3d(160, 320, 1, 1, 0, bias=False),
          nn.GroupNorm(40, 320),
          nn.ReLU(),
          nn.MaxPool3d((2, 2, 2)),
          
          
            )
        
    def activations_hook(self, grad):
        self.gradients = grad

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        
        x = self.conv3(x)
        h = x.register_hook(self.activations_hook)
        x = self.conv4(x)
        
        x = self.conv5(x)

        
        return x

    def get_activations_gradient(self):
        return self.gradients
    
    # method for the activation exctraction
    def get_activations(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        return x
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.encoder = Encoder()
        self.dropout = nn.Dropout(p = 0.2)
        self.convf = nn.Conv3d(320, 640, (2,2,2), bias=False, )
        self.relu = nn.ReLU()  
        self.fc1 = nn.Linear(640 * 4 * 2 * 2, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 40)
        self.fc4 = nn.Linear(40, 2)
        self.bn1 = nn.BatchNorm1d(120)
        self.bn2 = nn.BatchNorm1d(num_features=84)
        self.bn3 = nn.BatchNorm1d(num_features=40)

    def forward(self, x):
        x = self.encoder(x)
        x = self.relu((self.convf(x)))
        x = x.view(-1, 640 * 4 * 2 * 2)
        x = F.tanh(self.bn1((self.fc1(x))))
        x = self.dropout(x)
        x = F.tanh(self.bn2((self.fc2(x))))
        x = self.dropout(x)
        x = F.tanh(self.bn3((self.fc3(x))))

        return x
```

```python
def estrai_feature(dataloader):
    df = pd.DataFrame()
    if dataloader == 0:
        loader = train_loader
    else:
        loader = validation_loader
    outputs = []
    labels = []
    recs = []
    for i in range(0, 1):

        print("Training del modello")
        model_pre.eval()  # Set model to training mode

        for batch_idx, batch in enumerate(tqdm(loader)):

            if batch_idx < len(loader):


                inputs = batch['Lung'][DATA].to(dev)
                label = batch['label'].to(dev)
                id_paziente = batch['id']
                rec = batch['rec']
                print(int(id_paziente[0]))
                print(label.detach().cpu().numpy())

                output = model_pre(inputs)
                output_cpu = np.append(output.detach().cpu().numpy()[0], label.detach().cpu().numpy()[0])
                output_cpu = np.append(id_paziente[0], output_cpu)
                recs.append(rec[0])
                df = df.append([output_cpu])

    df = df.reset_index()
    df = df.drop(columns=['index'])
    df_rec = pd.DataFrame({'Rec': recs})   
    df = df.join(df_rec)
    if dataloader == 0:
        split = 'train'
    else:
        split = 'validation'
    df.to_csv('feature_estratte_{}.csv'.format(split))
    return df
```
